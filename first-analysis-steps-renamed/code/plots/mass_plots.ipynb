# Mass plots

Draw plots of mass spectra with toy $J/\psi \to \mu^{+}\mu^{-}$ and $B_{s}^{0} \to J/\psi\phi(1020)$ data.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
JPSI_MASS = 3097
JPSI_RESOLUTION = 7.5
WINDOW = (JPSI_MASS - 20*JPSI_RESOLUTION, JPSI_MASS + 20*JPSI_RESOLUTION)

bkg = np.random.uniform(*WINDOW, size=5000)
sig = np.random.normal(loc=JPSI_MASS, scale=JPSI_RESOLUTION, size=1000)
data = np.concatenate([sig, bkg])
```

```python
# One bin per 5 MeV
nbins = int(np.diff(WINDOW)/5)

fig, ax = plt.subplots()
ax.hist(data, range=WINDOW, bins=nbins, histtype='step')
ax.set_title(r'Invariant mass spetrum of $J/\psi \to \mu^{+}\mu^{-}$ candidates')
ax.set_xlabel(r'$m(\mu^{+}\mu^{-})$ [MeV/$c^{2}$]')
ax.set_ylabel(r'Candidates / 5 MeV/$c^{2}$')
fig.tight_layout()
fig.savefig('dimuon_mass.png', dpi=150)
```

```python
BS_MASS = 5367
BS_RESOLUTION = 10
WINDOW = (BS_MASS - 15*BS_RESOLUTION, BS_MASS + 25*BS_RESOLUTION)

bkg = np.random.exponential(scale=100, size=50000) + WINDOW[0]
sig = np.random.normal(loc=BS_MASS, scale=BS_RESOLUTION, size=5000)
data = np.concatenate([sig, bkg])
```

```python
# One bin per 5 MeV
nbins = int(np.diff(WINDOW)/5)

fig, ax = plt.subplots()
ax.hist(data, range=WINDOW, bins=nbins, histtype='step')
ax.set_title(r'Invariant mass spetrum of $B_{s}^{0} \to J/\psi\phi(1020)$ candidates')
ax.set_xlabel(r'$m(\mu^{+}\mu^{-}K^{+}K^{-})$ [MeV/$c^{2}$]')
ax.set_ylabel(r'Candidates / 5 MeV/$c^{2}$')
fig.tight_layout()
fig.savefig('jpsiphi_mass.png', dpi=150)
```
